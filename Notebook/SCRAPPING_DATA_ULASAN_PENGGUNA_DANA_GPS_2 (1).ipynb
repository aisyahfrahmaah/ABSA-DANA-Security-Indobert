```python
import pandas as pd
from google_play_scraper import Sort, reviews
from IPython.display import display
```

```python
from google_play_scraper import reviews_all
import pandas as pd
from datetime import datetime
```

```python
import time

# Asumsikan fungsi reviews dan enum Sort sudah diimpor dan dikonfigurasi
app_package = 'id.dana'
total_reviews = 3000
batch_size = 100

all_reviews = []
token = None

while len(all_reviews) < total_reviews:
    count = min(batch_size, total_reviews - len(all_reviews))
    result, token = reviews(
        app_package,
        lang='id',               # Bahasa Indonesia
        country='id',            # Negara Indonesia
        sort=Sort.NEWEST,        # Ambil dari yang terbaru
        count=count,
        continuation_token=token # Untuk mengambil batch selanjutnya
    )
    all_reviews.extend(result)
    print(f"Total ulasan terkumpul: {len(all_reviews)}")
    time.sleep(1)
```

```python
# Menampilkan beberapa data pertama
for review in all_reviews[:10]:  # Tampilkan 10 ulasan pertama
    print(review)
```

```python
import pandas as pd

# Mengonversi ulasan ke dalam DataFrame
df_reviews = pd.DataFrame(all_reviews)

# Menampilkan beberapa baris pertama dari DataFrame
print(df_reviews.head())
```

```python
# Simpan ke CSV
df_reviews.to_csv('nonfiltered_reviews.csv', index=False)
```

```python
# Cek kolom yang ada dalam DataFrame
print(df_reviews.columns)
```

```python
import pandas as pd
import re

# --- Cleaning Function ---
def clean_text(text):
    if pd.isnull(text):
        return ""
    text = text.lower()  # huruf kecil semua
    text = re.sub(r"http\S+|www\S+|https\S+", '', text, flags=re.MULTILINE)  # hapus URL
    text = re.sub(r'\@w+|\#','', text)  # hapus mention & hashtag
    text = re.sub(r"[^a-zA-Z\s]", '', text)  # hapus angka & tanda baca
    text = re.sub(r'\s+', ' ', text).strip()  # hapus spasi ganda
    return text
```

```python
# Terapkan cleaning ke kolom 'content'
df_reviews['clean_content'] = df_reviews['content'].apply(clean_text)
```
